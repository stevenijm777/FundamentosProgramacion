```python
# http://blog.espol.edu.ec/ccpg1001/1eva_iit2011_t3-parchis-2-fichas/
import random as rd
pistaP1 = ["-"]*50
pistaP2 = ["-"]*50
pistaP1[0] = "X"
pistaP2[0] = "Y"
print(pistaP1)
print(pistaP2)
terminado = False
turno = 0  
while not (terminado):
    dado = rd.randint(1, 6)
    turno += 1
    if turno % 2 != 0:
        posicion = pistaP1.index("X")
        if pistaP1[0] == "X":
            if dado == 5:
                pistaP1[1] = "X"
                pistaP1[posicion] = "-"
        else:
            new_posicion = posicion + dado
            if new_posicion < len(pistaP1):
                pistaP1[new_posicion] = "X"
                pistaP1[posicion] = "-"
            else:
                pistaP1[len(pistaP1)-1] = "X"
                pistaP1[posicion] = "-"
    else:
        posicion = pistaP2.index("Y")
        if pistaP2[0] == "Y":
            if dado == 5:
                pistaP2[1] = "Y"
                pistaP2[posicion] = "-"
        else:
            new_posicion = posicion + dado
            if new_posicion < len(pistaP2):
                pistaP2[new_posicion] = "Y"
                pistaP2[posicion] = "-"
            else:
                pistaP2[len(pistaP2) - 1] = "Y"
                pistaP2[posicion] = "-"
    ind_X = pistaP1.index("X")
    ind_Y = pistaP2.index("Y")
    if ind_X == ind_Y and ind_X != 0:
        if turno % 2 != 0:
            pistaP1[0] = "X"
            pistaP1[ind_X] = "-"
        else:
            pistaP2[0] = "Y"
            pistaP2[ind_Y] = "-"
    print(pistaP1)
    print(pistaP2)
    if ind_Y == len(pistaP2)-1 or ind_X == len(pistaP1)-1:
        break
ind_X = pistaP1.index("X")
ind_Y = pistaP2.index("Y")
if ind_Y == len(pistaP2)-1:
    print("GANO Y")
else:
    print("GANO X")
print(turno)
```
