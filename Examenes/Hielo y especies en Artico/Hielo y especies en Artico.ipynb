```python
import numpy as np

def cuadrantes(matriz): #(Cortesia del examen)
    f,c=matriz.shape
    f_mitad= f//2
    c_mitad=c//2
    Q1= matriz[:f_mitad, :c_mitad]
    Q2= matriz[:f_mitad, c_mitad:]
    Q3=matriz[f_mitad:, :c_mitad]
    Q4=matriz[f_mitad:, c_mitad:]
    return (Q1,Q2,Q3,Q4)

def poblacionEspecie(mAnimales, especie): #(Cortesia del examen)
    t=cuadrantes(mAnimales)
    totales=[]
    for Q in t:
        total=np.sum(Q==especie)
        totales.append(total)
    return tuple(totales)

dicEspecies={'0': 'Delfin',
             '1': 'Ballena',
             '2': 'Cocodrilo',
             '3': 'Venado',
             '4': 'Halcon',
             '5': 'Oso Polar',
             '6': 'Pez Globo'}
```

```python
def crearMatriz(nomArchivo):
  f = open(nomArchivo,'r')
  fila = int(f.readline()) 
  columna = int(f.readline())
  f.readline()
  # Crear Matrices
  M_hielo2009 = np.zeros((fila,columna))
  M_hielo2019 = np.zeros((fila,columna))
  M_especies2009 = np.zeros((fila,columna))
  M_especies2019 = np.zeros((fila,columna))
  for linea in f:
    anios,ind_fila,ind_col,hielo,especie = linea.strip().split(',') # 2009,0,9,1,5
    ind_fila = int(ind_fila)
    ind_col = int(ind_col)
    if anios == "2009":
      M_hielo2009[ind_fila, ind_col] = hielo
      M_especies2009[ind_fila,ind_col] = especie
    else:
      M_hielo2019[ind_fila,ind_col] = hielo
      M_especies2019[ind_fila,ind_col] = especie
  f.close()
  return (M_hielo2009,M_hielo2019,M_especies2009,M_especies2019)
```

```python
def densidadHielo(mHielo):
  Q1,Q2,Q3,Q4 = cuadrantes(mHielo)
  #iterar por elemento 
  Q = [Q1, Q2, Q3, Q4]
  dQ = [] #0.7 porq es ta es paralela a Q [0.7,0.8,0.5,0.4]
  for cuadrante in Q:
    celdas1 = [] #deben ser 1  5
    celdas0 = []  # 7
    filas = cuadrante.shape[0]
    cols  = cuadrante.shape[1]
    for i in range(filas): # 1
      for j in range(cols):# 0
        numero = cuadrante[i,j]
        if numero == 1:
          celdas1.append(numero) 
        else:
          celdas0.append(numero)
    densidad = len(celdas1)/ len(celdas1)+len(celdas0)
    dQ.append(densidad)
  return tuple(dQ)
```

```python
def especieDominante(mAnimales):
  Q1,Q2,Q3,Q4 = cuadrantes(mAnimales)
  Q = [Q1,Q2,Q3,Q4]
  aQ = []
  for cuadrante in Q:
    filas = cuadrante.shape[0]
    columnas = cuadrante.shape[1]
    l_elementos= [] # [4]
    l_cont = []     # [1]
    for i in range(filas):
      for j in range(columnas):
        valor_num = cuadrante[i,j]
        if valor_num not in l_elementos:
            l_elementos.append(valor_num)
            l_cont.append(1)
        else:
          ind = l_elementos.index(valor_num)
          l_cont[ind] += 1
      v_cont = np.array(l_cont)
      ind_mayor = v_cont.argmax()
      a_dominante = l_elementos[ind_mayor]
      aQ.append(a_dominante)
    return tuple(aQ)
```
