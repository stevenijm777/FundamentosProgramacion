```python
import random
import numpy as np

def generarDatos():
    random.seed(128)

    marcas = ['Kia', 'Chevrolet', 'Mazda', 'VW', 'Chery', 'Hyundai', 'Jeep', 'Toyota', 
              'Honda', 'Ford', 'Audi', 'Mercedez Benz', 'Renault', 'Citroen', 
              'Nissan', 'Great Wall', 'Seat', 'Changan', 'Skoda', 'Fiat']

    costa = ['Esmeraldas', 'Chone', 'Portoviejo', 'Manta', 'Bahia de Caraquez', 'Guayaquil', 'Salinas', 'Quevedo',
    'Babahoyo', 'Ventanas', 'La Libertad', 'Playas', 'Machala', 'Zaruma', 'Santa Rosa']

    sierra = ['Tulcan', 'Ibarra', 'Quito', 'Latacunga', 'Ambato', 'Riobamba', 'Baños', 'Azogues', 'Cuenca', 'Loja']

    oriente = ['Macas', 'Nueva Loja', 'Puyo', 'Tena', 'Zamora']

    ciudades = costa + sierra + oriente

    arr = np.zeros((len(marcas), len(ciudades)), int)

    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            marca = marcas[i]
            if marca in ['Mercedez Benz', 'Audi']:
                arr[i, j] = int(abs(random.gauss(50, 35)))
            elif marca in ['Kia', 'Chevrolet', 'Chery', 'Hyundai', 'Renault', 'Nissan', 'Great Wall', 'Changan']:
                arr[i, j] = int(abs(random.gauss(600, 200)))
            else:
                arr[i, j] = int(abs(random.gauss(200, 100)))

    return marcas, costa, sierra, oriente, arr
```

```python
marcas, costa, sierra, oriente, arr_ventas = generarDatos()
```

```python
#4. Mostrar cuál es la marca más vendida en cada ciudad
def vehiculos(arr_ventas):
  l_mas_vendidas = []
  col = arr_ventas.shape[1]
  for i in range(col):
    columna = arr_ventas[:,i]
    mayor = columna.argmax()
    vehi = marcas[mayor]
    l_mas_vendidas.append(vehi)
  return l_mas_vendidas
l_mas_vendidas =  vehiculos(arr_ventas)

def datos(l_mas_vendidas,ciudades):
  for k in range(len(ciudades)):
    ciu = ciudades[k]
    vehiculos = l_mas_vendidas[k]
    print('Ciudad: ',ciu)
    print('Marca mas vendida: ',vehiculos)
print(datos(l_mas_vendidas,ciudades))
```

```python
# una funcion recibe la matriz, las  lista marcas  y una tupla con varias marcas
# calcular el total de vantas de las marcas que estan en la tupla
marcas = ['Kia', 'Chevrolet', 'Mazda', 'VW', 'Chery', 'Hyundai', 'Jeep', 'Toyota', 
              'Honda', 'Ford', 'Audi', 'Mercedez Benz', 'Renault', 'Citroen', 
              'Nissan', 'Great Wall', 'Seat', 'Changan', 'Skoda', 'Fiat']
t_marcas = ('Kia','Ford','Renault', 'Changan')
def l_indices(listaGrande,listaPequeña):
    l_inds = []
    for elem in listaPequeña:
        ind = listaGrande.index(elem)
        l_inds.append(ind)
        
def total_marcas(arr_ventas, marcas,t_marcas):
    l_inds=l_indices(marcas,t_marcas)
    M2 = arr_ventas[l_inds,:]
    total_marcas = M2.sum(axis=1)
    return total_marcas
        

   
    
```
